```python
import sys
import PyPDF2
import PythonMagick
import os

def split2Png(pdf_file_name, out_dir):
    #counting page count of each pdf file 
    #pdffilename = os.path.join(f_path, '381.pdf')
    pdf_im = PyPDF2.PdfFileReader(pdf_file_name, "rb")
    npage = pdf_im.getNumPages()
    print(pdf_file_name + ": " +  str(npage))
    for p in range(npage):
        try:
            #setting the image dpi
            im = PythonMagick.Image()
            im.density('300')
            im.read(pdf_file_name + '[' + str(p) +']')
            im.write(out_dir+ '\\' + str(p+1)+ '.png')
        except (RuntimeError, TypeError, NameError):
            pass
 
        
```

```python
main_dir = r"D:\DataIncubator\PythonCodes\interactivepython\Al_Mojtama_Project\Matterials\AlMojtama"

years = os.listdir(main_dir)
for year in years:
    year_dir = os.path.join(main_dir, year)
    
    #if the file has already been splited exclude it
    pdfFileNames = [f for f in os.listdir(year_dir) if not os.path.isdir(os.path.join(year_dir, f.replace('.pdf', '')))]    
    for name in pdfFileNames:
        out_dir_name = name.replace('.pdf', '')
        out_dir  = os.path.join(year_dir, out_dir_name)
        pdf_file_name = os.path.join(year_dir, name)
        
        #create directory if not exists
        if not os.path.isdir(out_dir):
            os.mkdir(out_dir)
        
        # split pdf file into 
        try:
            split2Png(pdf_file_name , out_dir)
        except ValueError:
            pass
        os.remove(pdf_file_name)

```
