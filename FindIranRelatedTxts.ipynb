```python
import os
def getMagFolders(year = None):
    main_dir = r"D:\DataIncubator\PythonCodes\interactivepython\Al_Mojtama_Project\Matterials\AlMojtama"
    years = os.listdir(main_dir)
    dir_lst =list()
    if year == None:
        for year in years:
            year_dir =  os.path.join(main_dir , year )
            for mag_no in os.listdir(year_dir):
                mag_path = os.path.join(year_dir, mag_no)
                if os.path.isdir(mag_path):
                    dir_lst.append([mag_path, int(year), int(mag_no)])
    else:
        year_dir =  os.path.join(main_dir , str(year) )
        for mag_no in os.listdir(year_dir):
            mag_path = os.path.join(year_dir, mag_no)
            if os.path.isdir(mag_path):
                dir_lst.append([mag_path, year, int(mag_no)])
    return dir_lst
```

```python
def findPatternFrequency(txt, pattern):
    pos = 0
    count = 0
    txt_len = len(txt)
    while (pos > -1) and (pos < txt_len - 5):
        pos = txt[pos:].find(pattern)
        if pos > -1:
            count += 1
            pos = pos + 100000  
    return count
    
```

```python
import numpy as np
import pandas as pd

def magazineData(fyear, tyear):
    main_dir = r"D:\DataIncubator\PythonCodes\interactivepython\Al_Mojtama_Project\Matterials\AlMojtama"
    col_names =  ['year', 'mag_no', 'page_no', 'totoal_count', 'iran_count', 'almajoos_count', #'text', 
                  'fars_count', 'alAjam_count', 'alJomhouri_count']
    iranRelatedPages  = pd.DataFrame(columns = col_names)
    for year in range(fyear, tyear+1):        

        year_dir = os.path.join(main_dir, str(year))
        #consider all directories inside each year
        mag_dirs = [f for f in os.listdir(year_dir) if os.path.isdir(os.path.join(year_dir, f))]    
        for mag_dir in mag_dirs:
            page_files =[p for p in os.listdir(os.path.join(year_dir, mag_dir))
                             if os.path.isfile(os.path.join(year_dir, mag_dir, p))]

            for page_file in page_files:
                file_name = os.path.join(year_dir, mag_dir, page_file)

                filename, file_extension = os.path.splitext(file_name)
                txt = ''
                if file_extension == '.txt':
                    with open(file_name, 'rb') as f:
                        txt = f.read().decode('utf8')

                    totoal_count = len(txt.split(' '))
                    iran_count = findPatternFrequency(txt, u'إيران')
                    almajoos_count = findPatternFrequency(txt, u'المجوس')
                    fars_count = findPatternFrequency(txt, u'فارس ')
                    alAjam_count = findPatternFrequency(txt, u' العجم ')
                    alJomhouri_count = findPatternFrequency(txt, u'الجمهورية ايراني')
                    iranRelatedPages.loc[len(iranRelatedPages)] = [int(year), mag_dir, page_file.replace('.txt', ''), #txt, 
                                                                   totoal_count,iran_count, almajoos_count, fars_count, 
                                                                   alAjam_count, alJomhouri_count]

    return iranRelatedPages 
```

```python
for year in range(2016, 2018):
    name = 'almojtama_data_' + str(year) + '.txt'
    almojtama_data = magazineData(year, year)
    almojtama_data.to_csv(name, index = False )  #3:15
```
